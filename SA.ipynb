### Objective

```python
# Initialise 2D schubert function
schubert_2d = Schubert(2)
```

### Crude Method

```python
x0 = np.array([[1],[1]])
obj = schubert_2d.eval_obj
sa_2d = SA(obj, x0)
```

```python
x_opt, o_min, results_loc, results_obj, its = sa_2d.run()
```

```python
schubert_2d.plot(opt=results_loc, save='SA_scat_crude_a1', title='Simulated Annealing - Basic Method')
```

```python
plot_min_err(results_obj, title='2D Basic Simulated Annealing - {} evaluations'.format(its), save='crude_2d_minerr')
```

### Park's Method

```python
sa_parks = SA_parks(obj, x0)
```

```python
x_opt_p, o_min_p, results_loc_p, results_obj_p, its_p = sa_parks.run()
```

```python
schubert_2d.plot(opt=results_loc_p, save='SA_scat_parks', title='Simulated Annealing - Parks\' Method')
```

```python
plot_min_err(results_obj_p, title='2D Parks\' Method Simulated Annealing - {} evaluations'.format(its_p), save='parks_2d_minerr')
```
